# Data Visualization with cufflinks and plotly
*Cufflinks conncects plotly with pandas <br>
*Plotly is an interactive visualization library

```python
import pandas as pd
import numpy as np
from plotly.offline import download_plotlyjs, init_notebook_mode, iplot, plot
import cufflinks as cf
import seaborn as sns
import matplotlib as plt
```

```python
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
titanic= sns.load_dataset("titanic")
titanic.head()
```

```python
#It is showing the total number of people survived or died
titanic.groupby("survived").count()["alive"].iplot(kind="bar")
```

```python
#It is showing distribution of age
titanic["age"].iplot(kind="box")
```

```python
#Showing number of people boarded from different towns
titanic.groupby("embark_town").count()["embarked"].iplot(kind="bar")
```

```python
#Showin distribution of fare
titanic["fare"].iplot("hist", bins=15)
```
