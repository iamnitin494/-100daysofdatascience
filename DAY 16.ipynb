# Matplotlib cont.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#creating empty subplots using oops
fig, axes= plt.subplots(nrows=1, ncols=2)
```

```python
x= np.linspace(0.1,0.5,10)
x
```

```python
y=x**2
z=x**3
```

```python
fig, axes= plt.subplots(nrows=1, ncols=2) 
axes[0].plot(x,y)
axes[0].set_title("Square function")   #setting title for the graph
axes[0].set_xlabel("X-axis")           #setting x label
axes[0].set_ylabel("Y-axis")           #setting y label  

axes[1].plot(x,z)
axes[1].set_title("cube function")     #setting title for the graph
axes[1].set_xlabel("X-axis")           #setting x label
axes[1].set_ylabel("Y-axis")           #setting y label 
plt.tight_layout()                     #to prevent overlap ogf gaphs, we use tight_layout function.
```

```python
#changing the size of the plots
fig, axes= plt.subplots(nrows=1, ncols=2, figsize=(10,5))   #declaring size of the figure in figsize attribute.
axes[0].plot(x,y)
axes[0].set_title("Square function")
axes[0].set_xlabel("X-axis")
axes[0].set_ylabel("Y-axis")


axes[1].plot(x,z)
axes[1].set_title("cube function")
axes[1].set_xlabel("X-axis")
axes[1].set_ylabel("Y-axis")
plt.tight_layout()
```

```python
#saving your graphs
fig.savefig("my_functions.png", dpi=200)
```

```python
#Adding legends to your graphs

fig, axes= plt.subplots(nrows=1, ncols=2, figsize=(10,5))   #declaring size of the figure in figsize attribute.
axes[0].plot(x,y,label="Square function")
axes[0].plot(x,z,label="Cube function")
axes[0].legend()
axes[0].set_title("Multi function graph")
axes[0].set_xlabel("X-axis")
axes[0].set_ylabel("Y-axis")
```
