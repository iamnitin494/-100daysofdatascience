# Customization in matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.arange(0,10,1)
```

```python
x
```

```python
#changing color of the plot
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(10,6))
axes[0].plot(x,x**2, color="red")
axes[0].set_title("Square function")
axes[0].set_xlabel("x axis")
axes[0].set_ylabel("y axis")

axes[1].plot(x,x**3, color="black")
axes[1].set_title("cube function")
axes[1].set_xlabel("x axis")
axes[1].set_ylabel("y axis")

```

```python
#changing linewidth of the plot
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(10,6))
axes[0].plot(x,x**2, color="red", linewidth= 5)   
axes[0].set_title("Square function")
axes[0].set_xlabel("x axis")
axes[0].set_ylabel("y axis")

axes[1].plot(x,x**3, color="black",linewidth= 10)
axes[1].set_title("cube function")
axes[1].set_xlabel("x axis")
axes[1].set_ylabel("y axis")

```

```python
#changing linestyle of the plot
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(10,6))
axes[0].plot(x,x**2, color="red", linewidth= 5, linestyle='--')   
axes[0].set_title("Square function")
axes[0].set_xlabel("x axis")
axes[0].set_ylabel("y axis")

axes[1].plot(x,x**3, color="black",linewidth= 10, linestyle='dotted')
axes[1].set_title("cube function")
axes[1].set_xlabel("x axis")
axes[1].set_ylabel("y axis")

```

```python
#marking each data point of the plot 
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(10,6))
axes[0].plot(x,x**2, color="red", linewidth= 5, linestyle='--', marker="*", markersize=20, markerfacecolor="yellow")   
axes[0].set_title("Square function")
axes[0].set_xlabel("x axis")
axes[0].set_ylabel("y axis")

axes[1].plot(x,x**3, color="black",linewidth= 8, linestyle='dotted', marker="o", markersize=10,markerfacecolor="red")
axes[1].set_title("cube function")
axes[1].set_xlabel("x axis")
axes[1].set_ylabel("y axis")

```

```python
#setting limits for x axis and y axis 
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(10,6))
axes[0].plot(x,x**2, color="red", linewidth= 5, linestyle='--', marker="*", markersize=20, markerfacecolor="yellow")   
axes[0].set_title("Square function")
axes[0].set_xlabel("x axis")
axes[0].set_ylabel("y axis")
axes[0].set_xlim([3,8])

axes[1].plot(x,x**3, color="black",linewidth= 8, linestyle='dotted', marker="o", markersize=10,markerfacecolor="red")
axes[1].set_title("cube function")
axes[1].set_xlabel("x axis")
axes[1].set_ylabel("y axis")
axes[1].set_xlim([2,6])
```
