# Categorical Plots

```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
titanic= sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
#embark_town is a column which tells from where the people have boarded the ship
#countplot tells you the count of the total number of passengers from each place, they have boarded the ship from.
sns.countplot(titanic["embark_town"])
```

```python
#hue attribute is used to add more category in the graph
#This graph is telling about the number of people from different classes 
#such as first, second and third class who have boarded the ship from different locations
sns.countplot(titanic["embark_town"], hue = titanic["class"])
```

```python
#barplot can be thought of plotting a category according to some aggregate function
#This plot is showing the average age of males and females who have boarded the ship.
sns.barplot(x="sex", y="age", data=titanic, estimator= np.mean)
```

```python
#the boxplot is showing the quartiles of data and whiskers is showing the range of data
sns.boxplot(x="embark_town",y="age",hue='sex', data= titanic)
```

```python
#violinplot
sns.violinplot(x="embark_town",y="age",hue='sex', data= titanic)
```

```python
#catplot: It is the generalisation of each plot. You can plot different plots by specying the kind of plot in kind argument.
sns.catplot(x="embark_town",y="age",hue='sex', data= titanic, kind= "box")
```
