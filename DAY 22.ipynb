# Data Extraction from a pdf file using tabula
*tabula library is used to extract tables from pdf file <br>
*for installation: type <b>pip install tabula-py</b> in anaconda terminal

```python
from tabula import read_pdf
```

```python
#extracting tables from a pdf file
#by default it will extract the table from first page only
df= read_pdf(r"C:\\Users\\owner\\Desktop\\Food Calories List.pdf")
df
```

```python
#dropping the null values
df.dropna(axis=1, inplace=True)
```

```python
df
```

```python
#extracting the first page and getting the output in json format
df_json= read_pdf(r"C:\\Users\\owner\\Desktop\\Food Calories List.pdf", output_format='json')
df_json
```

```python
#extracting the table contents from third page
df_page3= read_pdf(r"C:\\Users\\owner\\Desktop\\Food Calories List.pdf",pages=3)
df_page3
```

```python
#extracting the table contents from all the pages at once
df_all= read_pdf(r"C:\\Users\\owner\\Desktop\\Food Calories List.pdf",pages="all", multiple_tables=True)
df_all
```

```python
#extracting the table contents from a page without any header
df_without_header= read_pdf(r"C:\\Users\\owner\\Desktop\\Food Calories List.pdf", pandas_options={"header":None})
df_without_header
```
