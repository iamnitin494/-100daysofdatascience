```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Matrix Plots

```python
titanic= sns.load_dataset("titanic")
```

```python
titanic.head()
```

```python
#displaying the nulll values in the dataset
titanic.isna().head(5)
```

```python
#total null values in the dataset
titanic.isna().sum()
```

```python
#finding the columns containing null values on the heatmap
sns.heatmap(titanic.isna())
```

```python
#corr() tells us the correlation between two quantitative variables
titanic.corr()
```

```python
#plotting correlation of each quantitative variable with other quantitative variables on a heatmap
plt.figure(figsize=(14,8))
sns.heatmap(titanic.corr(), annot= True)
```

# Gridplots

```python
#iris dataset is about comparison of different species of flowers
iris= sns.load_dataset("iris")
iris.head()
```

```python
#pairplot tells us the distribution and the scatterplots between all the quantitative variables in the dataset.
sns.pairplot(iris)
```

```python
#PAirGrid is used to customize your pairplot
grid= sns.PairGrid(iris)            
grid.map_diag(sns.distplot)     #customizing our diagonal plots
grid.map_upper(sns.scatterplot) #customizing our upper diagonal plots
grid.map_lower(sns.kdeplot)     #customizing our lower diagonal plots   
```

```python
#facetgrid is used to seggregate data on the basis of the categories of the two categories 
fgrid= sns.FacetGrid(data=titanic, row="survived", col="sex")
fgrid.map(sns.distplot,"fare", bins=1)
```
