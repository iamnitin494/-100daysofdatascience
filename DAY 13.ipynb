# REGEX
*Regular expressions(REGEX) in python<br>
*A way of describing pattern in a search string<br>
*It contains special characters to make the search in a string


#<b>Some keywords used in regex</b><br>
<b>\d</b> : digit (0-9) <br> <b>\w</b> : letter,digit or underscore<br> <b>\s</b> : whitespace character<br> <b>\D</b> : not a digit <br><b> \W </b>: not a word character<br><b> \S</b> : not a whitespace character<br><b> . </b>: any character except a line break



#<b>Quantifiers in regex</b><br>
<b>+</b> : one or more<br> <b>{x}</b>: exactly x times <br> <b>{3,5}</b> : 3 to 5 times <br> <b>{4,}</b>: 4 or more times <br> <b>*</b>: zero or more times <br> <b> ? </b>: once or none.

#<b>Character Classes</b><br>
<b> [ ] </b> : It will match one of the character in the bracket <br> <b> - </b> : range indicator <br> <b> [a-z]</b>: one of the character in therange from a to z <br> <b> [^x]</b>: one character that is not x <br> <b>[^a-z] </b>: one character that is not from range a - z
    

#<b>Anchors and boundaries</b><br>
<b>^</b>: start of string or line
<br><b>$ </b>: end of the string or line <br><b>\b </b> : word or boundary

```python
#importing regex library
import re
```

```python
#creating a regular expression object or pattern that will search a simple mobile number from a string
pattern= re.compile(r'\d{3} \d{3}-\d{4}')
```

```python
#searching a string with our pattern
#findall() will match all the matching patterns and returns them in a list
result = pattern.findall("My mobile number is 828 888-8580. Earlier it used to be 988 872-6627")
result
```

```python
#Creating a function that extracts a mobile number from a string.
def phone_extract(s):
    pattern= re.compile(r'\d{3}[-\s]\d{3}[-\s]\d{4}')
    return pattern.findall(s)
    
```

```python
#Testing the phone_extract function 
test_string= "My mobile number is 828 888 8580. Earlier it used to be 988-872-6627. My office number is 761 304 8584"
phone_extract(test_string)
```
