# Getting contents of a table from a website using Beautiful Soup

```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
#fetching the source code of url
data= requests.get("https://www.umggaming.com/leaderboards")
data.text
```

```python
soup= BeautifulSoup(data.text, "html.parser")
```

```python
print(soup.prettify)
```

```python
#just limiting the data we are interested in
table= soup.find("table", {"id":"leaderboard-table"})
table
```

```python
tbody= table.find("tbody")
tbody    
```

```python
#grabbing the positions of the users in a list
place= []
for tr in tbody.find_all("tr"):
    place.append(tr.find_all('td')[0].text.strip())
```

```python
place
```

```python
#grabbing the names of the users in a list
name=[]
for tr in tbody.find_all("tr"):
    name.append(tr.find_all('td')[1].text.strip())
```

```python
name
```

```python
#grabbing the points of the users in a list

points=[]
for tr in tbody.find_all("tr"):
    points.append(tr.find_all('td')[3].text.strip())
```

```python
points
```

```python
import pandas as pd
```

```python
#Creating a dataframe of the data we have fetched for the analysis
df= pd.DataFrame(list(zip(place,name,points)))
df= df.rename({0:"Position", 1:"Name", 2:"Points"},axis=1)
df
```
