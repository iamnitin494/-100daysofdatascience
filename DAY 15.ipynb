# Matplotlib
*widely used python library for plotting


```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#creating an array of numbers
x= np.linspace(1,5,50)
x
```

```python
#variable containing cube root of x
y= 1/x**2
y
```

```python
#plotting a cube root function between x, y
plt.plot(x,y)
```

```python
#adding labels axis and giving title to the plot
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Cube root function")
```

```python
#creating multiple plots in the same pane
#subplot(no_of_rows,no_of_columns,plot_number_you_are_referring_to)
z=x**2
plt.subplot(1,2,1)
plt.plot(x,z)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("square function")

plt.subplot(1,2,2)
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("cube root function")
```

```python
#plotting by using object oriented method
fig= plt.figure()   #creating a figure object
axes= fig.add_axes([0.1,0.1,0.5,0.5])   #calling method of that object
axes.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("cube root function")
```

```python
#adding two set of figures in one canvas
fig= plt.figure()   #creating a figure object
axes1= fig.add_axes([0.1,0.1,0.5,0.5])    
axes2= fig.add_axes([0.8,0.1,0.5,0.5])   

axes1.plot(x,y)
axes2.plot(x,z)
axes1.set_title("cube root function")
axes2.set_title("square function")
```
